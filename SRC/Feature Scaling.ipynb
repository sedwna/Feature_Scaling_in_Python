```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.preprocessing as pp
```

```python
Features = ['Age', 'Weight', 'Height']
X = np.array([[23, 154.25, 67.75],
             [22, 173.25, 72.25],
             [22, 154, 66.25],
             [26, 184.75, 72.25],
             [24, 184.25, 71.25],
             [24, 210.25, 74.75],
             [26, 181, 69.75],
             [25, 176, 72.5],
             [25, 191, 74],
             [23, 198.25, 73.5]])
```

```python
for i, f in enumerate(Features):
    plt.subplot(1, len(Features), i+1)
    plt.hist(X[:, i], bins=4)
    plt.xlabel(f)
    plt.ylabel('N')

plt.show()
```

```python
Scaler1 = pp.MinMaxScaler()
X1 = Scaler1.fit_transform(X)
```

```python
for i, f in enumerate(Features):
    plt.subplot(1, len(Features), i+1)
    plt.hist(X1[:, i], bins=4)
    plt.xlabel(f)
    plt.ylabel('N')

plt.show()
```

```python
Scaler2 = pp.StandardScaler()
X2 = Scaler2.fit_transform(X)
```

```python
for i, f in enumerate(Features):
    plt.subplot(1, len(Features), i+1)
    plt.hist(X2[:, i], bins=4)
    plt.xlabel(f)
    plt.ylabel('N')

plt.show()
```

```python
M = np.mean(X2, axis=1)
V = np.var(X2, axis=1)

print(f'{M = }')
print(f'{V = }')
```
